```python
mweb={1:[2,3],2:[4,5],3:[5,6],4:[1],5:[7,8],6:[1],7:[1],8:[1]}
T= array([[1./len(mweb[i]) if m in mweb[i] else 0. for m in mweb] for i in mweb])
around(reduce(dot,[T]*25),3)
p=0.
n=len(T)
Tp=p*ones([n,n])/n + (1-p)*T
linalg.matrix_power(Tp,50)
```
